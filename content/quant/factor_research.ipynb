### Fill Missing Vlues

#### Idea

We use an algorithm like ***K nearest neighbors (KNN)*** to find similar stocks and fill missing values using their mean:
1. standardize values of each column (within a month's range) in order not to let large-scale columns dominate
2. Compute *euclidean distance* as a measure of similarity
3. Find $k$ (we use $k=5$) stocks with the smallest distance
4. Compute the mean of therir value on missing columns.

#### Define Functions

```python
def fill_row(row, others, n_neighbor=5):
    """This is actually KNN algo.
    Parameters
    -----------
    row is a Series, others is a DataFrame. """
    missing_col_bool = np.isnan(row)
    valid_col_bool = np.logical_not(missing_col_bool)
    
    dist = np.sum((row[valid_col_bool] - others[:, valid_col_bool])**2, axis=1)# 1D array, length = len(others)
    argsort = np.argsort(dist) # argsort is the index which can sort the array
    
    k_nearest = others[argsort[: n_neighbor], :]
    mean = np.mean(k_nearest, axis=0)
    
    row[missing_col_bool] = mean[missing_col_bool]
    return row

def process_missing(mat, n_neighbor=5):
    """mat is a 2D np.ndarray. """
    # get index
    is_missing = np.isnan(mat)
    missing_rows_bool = np.any(is_missing, axis=1)
    
    # deal with corner cases
    if np.any(missing_rows_bool) == False:
        return mat
    
    # standardization
    means, stds = np.nanmean(mat, axis=0), np.nanstd(mat, axis=0)
    mat_std = (mat - means) / stds
    
    # assigning the result of advanced indexing to a variable creates a copy of an array
    rows_missing = mat_std[missing_rows_bool, :]
    rows_valid = mat_std[np.logical_not(missing_rows_bool), :]
    
    # process each row with missing values
    mat_std[missing_rows_bool, :] = np.apply_along_axis(
        lambda row: fill_row(row, rows_valid, n_neighbor), axis=1, arr=rows_missing)
    
    # transform back
    return mat_std * stds + means


tmp = raw.loc['2008-12-31', :].copy()
tmp2 = tmp.copy()
tmp2.loc[:, 'TotalRet': 'Risk12'] = process_missing(tmp2.loc[:, 'TotalRet': 'Risk12'].values)
#tmp.describe()
#tmp2.describe()
```

**Strength of the function**:
1. It operates on `np.ndarray` instead of `pd.DataFrame` , which is way faster. It is also fully vectorized;
2. It will only standardize continuous columns, leaving discrete columns remain the same.

## Data Preprocessing: Extreme Values (outlier)

To detect extreme values, we should first calculate statistics such as median. Therefore, we **must first fill NaN values**.

### Method

Usually, we use **mean** and **standard deviation** to detect outliers. However it is **not robust**.

Here we use a more robust method, which take advantage of ***Median Absolute Deviation**. For details see:

Leys, C.; et al. (2013). "Detecting outliers: Do not use standard deviation around the mean, use absolute deviation around the median". *Journal of Experimental Social Psychology*. 49: 764–766. doi:10.1016/j.jesp.2013.03.013

```python
def process_outlier(arr):
    CONST_RATIO = 1.483
    mid = np.median(arr)
    mad = np.median(np.abs(arr - mid))
    mad_e = mad * CONST_RATIO
    
    lower, upper = mid - 3*mad_e, mid + 3*mad_e
    
    arr[arr < lower] = lower
    arr[arr > upper] = upper
    
    return arr

tmp = raw.loc['2008-12-31', :].copy()
tmp.loc[:, 'TotalRet': 'Risk12'] = process_missing(tmp.loc[:, 'TotalRet': 'Risk12'].values)
tmp2 = tmp.copy()
#print(tmp2.sum())
tmp2.loc[:, 'Risk1': 'Risk12'] = np.apply_along_axis(process_outlier, axis=0, arr=tmp2.loc[:, 'Risk1': 'Risk12'].values)
#print(tmp2.sum())
```

**Strength of the function**:

It only operates on 1D `np.ndarray`, which is simple and easily maintain. We just apply it simutaneously on multiple columns.

### Normality Test for Risk Loadings

Question we want to tackle is: Do we need **standardization** or **transformation**?

```python
def norm_test(data, axis=0, nan_policy='propagate'):
    st_skew, pv_skew = scst.skewtest(data, axis=axis, nan_policy=nan_policy)
    st_kurt, pv_kurt = scst.kurtosistest(data, axis=axis, nan_policy=nan_policy)
    st_omni, pv_omni = scst.normaltest(data, axis=axis, nan_policy=nan_policy)
    
    round_num = 5
    pv_skew, pv_kurt, pv_omni = round(pv_skew, round_num), round(pv_kurt, round_num), round(pv_omni, round_num)
    
    arr = np.array([[st_skew, pv_skew], [st_kurt, pv_kurt], [st_omni, pv_omni]])
    df = pd.DataFrame(index=['Skewtest', 'Kurtosistest', 'Omnitest'], columns=['Statistic', 'P value'], data=arr)
    df.index.name = 'Test Name'
    return df
```

```python
month = raw.loc['2008-12-31', :].copy()
month.loc[:, 'TotalRet': 'Risk12'] = process_missing(month.loc[:, 'TotalRet': 'Risk12'].values)
month.loc[:, 'Risk1': 'Risk12'] = np.apply_along_axis(process_outlier, axis=0, arr=month.loc[:, 'Risk1': 'Risk12'].values)

for i in range(1, 14):
    tmp = month.loc[:, 'Risk' + str(i)]
    sns.distplot(tmp)
    print(norm_test(tmp))
    plt.show()
```

Most risk factor's distribution are positive skewed and kurtosised.

## Regression

### Mathematical Model

$$
r_n = \sum_{k} X_{nk} f_k + u_n,
$$
where $X_{nk}$ is the exposure of stock $n$ to factor $k$ , $f_k$ is the return to the factor, and $u_n$ is the stock specific return.
 
***Reference***: Barra's United States Equity Model Version 4 (USE4)

Note:
- About **intercept** term: It represents the average of specific returns of all stocks in the pool. It can be non-zero. Therefore we will fit the intercept in regression.
- About **significance test** of coefficients: The significance test makes no sense here, since **return of a factor can be zero**. (Unlike ordinary regression contexts, where we may drop a variable when its coeffienct is not significant;

### Data Preparation

### Heteroscedasticity of OLS

We detect heteroscedasticity by computing **correlation** between OLS residuals and columns we have.

We do not use statistical tests for heteroscedasticity because we would like to know can we **solve it using information (columns) that we already have**.

### WLS

If we found $residual \propto x$:

We kone range of $residual \propto \sigma$

so:
$$
{variance}^{-1/2} = {\sigma}^{-1} = x^{-1}
$$

According to `sklearn`'s implementation, `sample_weight` should be $1 / \Sigma$, where $\Sigma$ is diagnal covariance matrix.

Here we combine 1/log(`MktCap`) and `Volatility` as an approximation of the standard deviation ($\sigma$) of $u_n$ in our model.

Note we **do not use residual of OLS** as an approximation because that is not stable and can cause **over-fitting**.

```python
def calc_sigma_arr(df):
    """df is a month DataFrame, return 1D array. """
    return (1.
            / np.log(1 + df.loc[:, 'MktCap'].values)
            * df.loc[:, 'Volatility'].values
           )
```

```python
sigma_arr = calc_sigma_arr(month)
```

```python
rgsr = skllinear.LinearRegression(fit_intercept=True)
rgsr.fit(X, Y,
         sample_weight=np.power(sigma_arr, -2)
        )
print("Rsqured = {:6.4f}".format(rgsr.score(X, Y)))
print("intercept = {:6.4f}".format(rgsr.intercept_))
rgsr.coef_
```

We see that *Heteroscedasticity* has been greatly reduced.

## Fama French

### Portfolio Construction Using Optimization: target

$X$ is loading matrix, $\boldsymbol{\alpha}$ is coef for each stock, we want construct a portfolio **that only has exposure to risk $i$**. 

Since portfolio loading is $X^T \boldsymbol{\alpha}$, **our target is**:  $X^T \boldsymbol{\alpha} = [0, 0, \dots, 1, 0, 0, \dots, 0]$, which is a vector with only one non-zero term at position $i$.

To be clear, we can divide $X$ into 2 matrix, $X_1$ and $X_2$. $X_2$ is the loading of risk factor $i$ and $X_1$ is the loading of other risk factors. Then we want: 
$$
\begin{cases}
X_1^T \boldsymbol{\alpha} &= \mathbf{0} \qquad (1)\\
X_2^T \boldsymbol{\alpha} &= 1 \qquad (2)
\end{cases}
$$

Then, according to our target, we form it into a **standard optimization problem** and use `scipy.optimize` package to solve it.

### Our first (unsuccessful) model

#### Model

Eq.(1) is linear constraints, but if we solve our problem under these constraints, we might not find a feasible solution. Thus we consider **minimize the sum of square of loadings** on other risk factors, given Eq.(2):
$$
\begin{aligned}
\min_{\boldsymbol{\alpha}} ~& (X_1^T \boldsymbol{\alpha})^T X_1^T \boldsymbol{\alpha}\\
\textrm{s.t.} ~& X_2^T \boldsymbol{\alpha} = 1.
\end{aligned}
$$
i.e.
$$
\begin{aligned}
\min_{\boldsymbol{\alpha}} ~& \boldsymbol{\alpha}^T (X_1 X_1^T) \boldsymbol{\alpha}\\
\textrm{s.t.} ~& X_2^T \boldsymbol{\alpha} = 1,
\end{aligned}
$$
which is a **quadratic optimization** problem.

#### Implementation

When calling `scipy.optimize.minimize`, I choose the `SLSQP` solver which has $O(n^3)$ time complexity. 

Because it takes several minutes to solve this optimization. I implement it in a separate Python file, see `optimize_model1.py`

#### A Severe Problem

Having applied our implementation of model1 on our data, we found that although our constraints and objective are perfectly achieved, there is a severe problem:

- **PHENOMENON**: The optimization results (**weight** of each stock in portfolio) for **different** risk factors only have **slight difference**. Thus return of portfolio for different risk factors are almost the same. 
- **ANALYSIS**: The **number of coefficients** (size of stock pool) is too large. They just need slight adjustment to adapt to different constraints (we only have one linear constraint). Therfore when seeking the minimum of objective function, they fall into **the same local minimum**.
- **CONCLUSION**: Optimization results are not reliable. We need to modify our model.

### Our second model: idea

#### Model

This is inspired by the *double-sort* method.

1. **Sort risk matrix** $X$ according to the **value of risk factor $i$**;
2. Long top 30% stocks and short bottom 30% stocks using **unequal weights**;
3. The weights are determined s.t. **portfolio's exposure on other risk factors equal zero**;
4. We leave portfolio's **exposure on risk factor $i$ uncontrolled**. We can divide this exposure when calculating return of risk factor $i$.

State our idea in the **standard form** (suppose $n'$ is the number of 30% stocks and $n$ is the number of all stocks):
$$
\begin{aligned}
\min_{\boldsymbol{\alpha}} ~& \boldsymbol{\alpha}^T (X_1 X_1^T) \boldsymbol{\alpha}\\
\textrm{s.t.} ~& \alpha_1, \alpha_2, \dots, \alpha_{n'} \ge 1\\
               & \alpha_{n'+1}, \alpha_{n'+2}, \dots, \alpha_{n} \le -1\\,
\end{aligned}
$$

#### Implementation

Again, we implement this model in a separate file. See `optimize_ret.py`.

#### Test Results

- Our objective function $\boldsymbol{\alpha}^T (X_1 X_1^T) \boldsymbol{\alpha}$ can be minimized to **less than `1e-6`**.
- Our constraint can **all** be achieved.
- The return of risk factors we computed are **close to those computed in regression**.
